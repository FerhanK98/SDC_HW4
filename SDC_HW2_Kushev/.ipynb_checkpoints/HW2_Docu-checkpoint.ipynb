## Data: 
https://www.ecdc.europa.eu/en/publications-data/data-covid-19-vaccination-eu-eea

```python
import pandas as pd
df = pd.read_csv('data.csv') # https://www.ecdc.europa.eu/en/publications-data/data-covid-19-vaccination-eu-eea 
df.head()
```

### Only a subset of the data is being used

```python
df = df[(df.TargetGroup != 'Age<18') & 
        (df.TargetGroup != 'AgeUNK') & 
                (df.TargetGroup != 'ALL') & 
        (df.TargetGroup != 'HCW') & 
        (df.TargetGroup != 'LTCF') & 
        (df.TargetGroup != '1_Age60+') & 
        (df.TargetGroup != '1_Age<60') &
        (df.YearWeekISO.str.startswith('2022-W22'))] # from June 2022 until November 2022
```

## Dashboard
- consists of two filters (single select)
 - First filter is about selecting First Dose, Second Dose or both to be displayed distributed by Age Group (TargetGroup)
 - Second filter is a country filter including an ALL value
- For filter options "First Dose" and "Second Dose" a simple bar chart is shown (How many people in the ages from 0 to 4 received a first dose vaccine from June 2022 until November 2022?)
- For both options a scatter plot is shown, where x is the first dose and y the second dose, again distributed by Age Group
- A HTML formatted number is also displaying the population of the selected country (country filter)
- Additionally under the bar chart / scatter plot there are plots about the distribution of vaccines, which are also filterable
- In the bottom there is a simple FastAPI calculation, which calcualtes: A selected number x is how much percent of a selected number y (x/y)
